#Practice Task
Create a file named myinfo.txt
Write your name, department, and course
Read and display the contents using read() and readline()
Read and display the contents using read() and readline() using with statement

```python
file= open("myinfo.txt","w")
file.write("ZAHRA RAHIMI\n")
file.write("I AM IN UG 1.2 \n ")
file.write("I AM IN COMPUTER SCIENCE DEPARTMENT.\n")
file.close()


file= open("myinfo.txt","r")
print(file.read())

print("using readline():")
with open("myinfo.txt","r") as i:
  print(i.readline())

with open("myinfo.txt","r") as f:
  print(f.readlines())

with open("myinfo.txt","r") as d:
  a= d.readlines()
  print(a[2])
```
